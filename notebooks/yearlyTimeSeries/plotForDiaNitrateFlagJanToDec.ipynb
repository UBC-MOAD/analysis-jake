```python
diff_cum_sum_nitrate = nitrate_sum * 86400 * 30.5
diff_cum_sum_diatoms = diatoms_sum * 86400 * 30.5
diff_cum_sum_flagellates = flagellates_sum * 86400 * 30.5
```
